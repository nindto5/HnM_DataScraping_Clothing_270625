**DATA REDUCTION CLOTHING LADIES USA**

```python
import pandas as pd

df = pd.read_csv('HnM_CLOTHING_LADIES_USA_010725.csv')
with pd.option_context('display.max_rows', 5, 'display.max_columns', None): 
    display(df)
```

```python
columns_to_display = ['sku', 'id', 'name', 'final_price_value', 'final_price_currency', 'supplier_country']

df_selected_columns = df[columns_to_display]
display(df_selected_columns)
```

```python
output_filename = 'HnM_CLOTHING_LADIES_USA_REDUCED.csv'
df_selected_columns.to_csv(output_filename, index=False)

print(f"Successfully saved selected columns to '{output_filename}'")
```
